```python
import pandas as pd 
customer1=pd.read_csv("D:/customers-1000.csv",header=0,index_col=0)
customer1
```

```python
people_rawdata1=pd.read_csv("D:/people-1000.csv",header=0,index_col=0)
people_rawdata1
```

```python
# To join the tables as join query merge both tables inner join by default
cust_people=pd.merge(customer1,people_rawdata1,on='Index')
cust_people
```

```python
# where query 


sort_query=cust_people.sort_values('Subscription Date') 
sort_query
```

```python
# And Query 

and_query=cust_people[(cust_people["Sex"]== "Male") & (cust_people["Country"]== "China")]
and_query
```

```python
# count and groupby query 
groupby_custpeople=cust_people.groupby('Country').count()
groupby_custpeople
```

```python
size_custpeople=cust_people.groupby('Job Title').size()
size_custpeople
```
